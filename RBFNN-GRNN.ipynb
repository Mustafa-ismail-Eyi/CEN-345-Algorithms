```python
import warnings
warnings.filterwarnings("ignore")

# keras is a deep learning library under the tensorflow
import keras
import tensorflow as tf
from keras.layers import Dense, Input, Dropout
from keras.models import Sequential
# sklearn for machine learning and data processing
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
# numpy for numerical operators
import numpy as np
# this package under this line for building a rbf layer
from rbf_keras import rbflayer
# matplotlib for plotting graphs 
from matplotlib import pyplot as plt
```

## Data & Data Preprocessing

```python
# loading the data, the data is iris data set which is populer
# this data set has 3 kind of iris flower and their length and widths of petals and sepals 
# data has pre-built dataset which is in sklearn library
data = datasets.load_iris()

# loading length and widths of sepals and petals  
X = data.data

# loading the kinds which are actually Iris-setosa, Iris-versicolor, Iris-virginica
# But in this data set it is already one-hot encoded and it seems like 0-1-2
y = data.target

# for the machine learning algorithm we need the split the data into train and test
# for better learning it is better to scaling up the data between 0-1
X_train, X_test, y_train, y_test = train_test_split(preprocessing.minmax_scale(X),(y.reshape((-1,1))), 
                                                    test_size=0.25)
```

## Building RBFNN model & Training & Analysis

```python
# Our model is a sequential model which take processes in a sequence
RBFNN_model = Sequential()

#x = Input(shape=X_train.shape)
#RBFNN_model.add(Input(X_train.shape))
#RBFNN_model.add(Dense(X_train.shape[1],input_shape=[X_train.shape[1]],activation='sigmoid'))

# This is the RBF layer
RBFNN_model.add(rbflayer.RBFLayer(2,
                       initializer=rbflayer.InitCentersRandom(X_train),
                       betas=0.5,
                       input_shape=(X_train.shape[1],)
                       ))
#RBFNN_model.add(Dense(5))
#RBFNN_model.add(Dropout(0.3))


# This is the last output layer 
RBFNN_model.add(Dense(1,activation='sigmoid'))
# Our main purpose is clustering that's why we choosed the parameters appropriately
RBFNN_model.compile(loss='mean_squared_error', optimizer ='adam', metrics=['categorical_accuracy'])
```

```python
# Let's take a quick watch at out model.
RBFNN_model.summary()
```

```python
# Let's train out model
history = history = RBFNN_model.fit(
    X_train, y_train,
    batch_size=8,
    epochs=500, # shows how many times ANN will be trained
    validation_data=(X_test,y_test), # for testing
    verbose=1, # shows the outputs
)
# RBFNN is very powerful model. It gives a very good performance in a short train time
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# We choosed train steps very large because to show the graph below
```

## Libraries for GRNN

```python
# loading the package of GRNN
from pyGRNN import GRNN

# for finding the optimal parameters
from sklearn.model_selection import  GridSearchCV

# for getting the mean absolute error 
from sklearn.metrics import mean_squared_error as MSE
```

```python
# creating instance of object
GRNN = GRNN()

# setting up the parameters for GridSearchCV
params_GRNN = {'kernel':["RBF"],
             'sigma' : list(np.arange(0.1, 4, 0.01)),
             'calibration' : ['None']
              }

# GridSearchCV tries to find best parameters for above
grid_GRNN = GridSearchCV(estimator=GRNN,
                       param_grid=params_GRNN,
                       scoring='neg_mean_squared_error',
                       cv=5,
                       verbose=1
                       )
# training the model
grid_GRNN.fit(X_train, y_train.ravel())

# From models created by GridSearchCV we get the best options we have
best_model = grid_GRNN.best_estimator_

# with the best model we try to predict our test data
y_pred = best_model.predict(X_test)

# to calculate performance measure the error with MSE
mse_GRNN = MSE(y_test, y_pred)
```
